##1Ô∏è‚É£ Deskripsi Kasus Penggunaan
###üìå Masalah

Dataset Iris memiliki 4 fitur numerik (sepal length, sepal width, petal length, petal width). Data berdimensi tinggi sulit untuk divisualisasikan dan dipahami pola hubungannya secara langsung.

###üìå Mengapa Dimensionality Reduction Dibutuhkan

Dimensionality reduction digunakan untuk:

Memvisualisasikan data berdimensi tinggi ke 2D

Melihat pola dan cluster antar kelas

Membantu eksplorasi data sebelum modeling


##2Ô∏è‚É£ Dataset

- Nama: Iris Dataset

- Sumber: sklearn.datasets.load_iris

- Jumlah data: 150 sampel

- Fitur: 4 fitur numerik

- Label: 3 kelas bunga Iris

##3Ô∏è‚É£ Implementasi di Python (Notebook)

```python
#Import Library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
#Load Dataset
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names
```

```python
#Standardisasi Data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

##4Ô∏è‚É£ Penerapan PCA

```python
#Reduksi ke 2 Dimensi
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
#Visualisasi PCA
plt.figure()
sns.scatterplot(
    x=X_pca[:,0],
    y=X_pca[:,1],
    hue=y,
    palette="deep"
)
plt.title("Visualisasi PCA 2D")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```

###üìåPenjelasan PCA

- PCA mempertahankan variasi terbesar dalam data

- Komponen utama adalah kombinasi linier fitur asli

- Terlihat bahwa kelas Setosa terpisah jelas, sedangkan dua kelas lainnya agak tumpang tindih

##5Ô∏è‚É£ Penerapan t-SNE

```python
#Reduksi ke 2 Dimensi
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
X_tsne = tsne.fit_transform(X_scaled)
```

```python
#Visualisasi t-SNE
plt.figure()
sns.scatterplot(
    x=X_tsne[:,0],
    y=X_tsne[:,1],
    hue=y,
    palette="deep"
)
plt.title("Visualisasi t-SNE 2D")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.show()
```

##üìå Penjelasan t-SNE

- t-SNE fokus menjaga kedekatan lokal

- Cluster antar kelas terlihat lebih jelas dan terpisah

- Cocok untuk eksplorasi struktur cluster non-linear

##6Ô∏è‚É£ Analisis Perbandingan PCA vs t-SNE
###üîç Perbedaan Hasil

- PCA

  - Linear

  - Menjaga variasi global

  - Cocok untuk preprocessing sebelum modeling

- t-SNE

  - Non-linear

  - Sangat baik untuk visualisasi cluster

  - Tidak cocok untuk modeling lanjutan

### Metode yang Lebih Sesuai

Untuk kasus visualisasi dan eksplorasi cluster, t-SNE lebih sesuai karena mampu menampilkan pemisahan cluster dengan lebih jelas dibanding PCA.

##Kesimpulan
Dimensionality reduction menggunakan PCA dan t-SNE terbukti efektif untuk membantu memahami data berdimensi tinggi. PCA mampu mereduksi dimensi dengan mempertahankan variasi global data sehingga cocok digunakan sebagai tahap awal analisis atau preprocessing sebelum pemodelan. Sementara itu, t-SNE lebih unggul dalam menampilkan struktur lokal dan pemisahan cluster, sehingga sangat efektif untuk keperluan visualisasi dan eksplorasi pola data.

Berdasarkan hasil visualisasi, t-SNE lebih sesuai untuk kasus visualisasi dan eksplorasi cluster, sedangkan PCA lebih tepat untuk analisis fitur dan reduksi dimensi sebelum modeling. Dengan demikian, pemilihan metode dimensionality reduction perlu disesuaikan dengan tujuan analisis yang ingin dicapai.
